```python
import requests
import json

# Define the URL of the Ollama API
url = "http://localhost:11434/api/generate"

# Define the headers for the request
headers = {
    "Content-Type": "application/json"
}

# Define the data for the request
data = {
    "model": "codestral",
    "prompt": "donne moi la requete sql seulement pour avoir tous les nom des personnes dont l'age est supérieur a 30",
    "stream": False
}

# Convert the data to JSON
data_json = json.dumps(data)

# Send the POST request
response = requests.post(url, headers=headers, data=data_json)

# Print the response
print(response.json())
```

```python
from llama_cpp import Llama

# Chargez le modèle
llm = Llama(model_path="./mistral-7b-instruct-v0.1.Q5_K_S.gguf" ,n_gpu_layers=0)

# Définissez la fonction pour convertir une phrase en langage naturel en requête SQL
def convert_to_sql(input_text):
    # Ajoutez une instruction spécifique pour générer la requête SQL
    input_text = "Convert the following natural language query into a SQL query: " + input_text

    # Générez la requête SQL
    output = llm(input_text, max_tokens=100)

    # Extrayez la requête SQL de la sortie
    sql_query = output["choices"][0]["text"].strip()

    return sql_query

# Testez la fonction de conversion
input_text = "Quel est le nom du client qui a passé la commande 123 ?"
sql_query = convert_to_sql(input_text)
print(sql_query)
```
